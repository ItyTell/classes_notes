```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.asarray(Image.open('cameraman.tif'))
```

```python
img.shape
```

```python
plt.imshow(img, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
def mashtab(img, a = 0, b = 255):
    return (img - img.min()) / (img.max() - img.min()) * (b - a) + a
```

```python
log_img = np.log(img + 1)
log_img = mashtab(log_img)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(log_img, cmap=plt.get_cmap('gray'), vmin=0,vmax=255)
axs[0].set_title('Log image')
axs[0].axis('off')
axs[1].imshow(img, cmap=plt.get_cmap('gray'), vmin=0,vmax=255)
axs[1].set_title('Original image')
plt.axis('off')
plt.savefig("img_1")
plt.show()
```
